# lasagne
* lasagne is a library for neural network building and training
* it's a low-level library with almost seamless integration with theano


````python
import lasagne
import theano
import theano.tensor as T

input_X = T.tensor4("X")

#input dimention (None means "Arbitrary")
input_shape = [None,3,32,32]

target_y = T.vector("target Y integer",dtype='int32')
````

Defining network architecture

````python
#Input layer (auxilary)
input_layer = lasagne.layers.InputLayer(shape = input_shape,input_var=input_X)

#fully connected layer, that takes input layer and applies 50 neurons to it.
# nonlinearity here is sigmoid as in logistic regression
# you can give a name to each layer (optional)
dense_1 = lasagne.layers.DenseLayer(input_layer,num_units=100,
                                   nonlinearity = lasagne.nonlinearities.sigmoid,
                                   name = "hidden_dense_layer")

#fully connected output layer that takes dense_1 as input and has 10 neurons (1 for each digit)
#We use softmax nonlinearity to make probabilities add up to 1
dense_output = lasagne.layers.DenseLayer(dense_1,num_units = 10,
                                        nonlinearity = lasagne.nonlinearities.softmax,
                                        name='output')
````

````python
#network prediction (theano-transformation)
y_predicted = lasagne.layers.get_output(dense_output)
````

````python
#all network weights (shared variables)
all_weights = lasagne.layers.get_all_params(dense_output,trainable=True)
print all_weights
````

### Than you could simply
* define loss function manually
* compute error gradient over all weights
* define updates
* But that's a whole lot of work and life's short
  * not to mention life's too short to wait for SGD to converge

Instead, we shall use Lasagne builtins

````python
#Mean categorical crossentropy as a loss function - similar to logistic loss but for multiclass targets
loss = lasagne.objectives.categorical_crossentropy(y_predicted,target_y).mean()

#prediction accuracy (WITH dropout)
accuracy = lasagne.objectives.categorical_accuracy(y_predicted,target_y).mean()

#This function computes gradient AND composes weight updates just like you did earlier
updates_sgd = lasagne.updates.adam(loss, all_weights,learning_rate=0.01)
````

````python
#function that computes loss and updates weights
train_fun = theano.function([input_X,target_y],[loss,accuracy],updates= updates_sgd)
````

````python
#deterministic prediciton (without dropout)
y_predicted_det = lasagne.layers.get_output(dense_output,deterministic=True)

#prediction accuracy (without dropout)
accuracy_det = lasagne.objectives.categorical_accuracy(y_predicted_det,target_y).mean()

#function that just computes accuracy without dropout/noize -- for evaluation purposes
accuracy_fun = theano.function([input_X,target_y],accuracy_det)
````

### That's all, now let's train it!
* We got a lot of data, so it's recommended that you use SGD
* So let's implement a function that splits the training sample into minibatches

````python
# An auxilary function that returns mini-batches for neural network training
# https://github.com/Lasagne/Lasagne/blob/master/examples/mnist.py

#Parameters
# X - a tensor of images with shape (many, 3, 32, 32), e.g. X_train
# y - a vector of answers for corresponding images e.g. Y_train
#batch_size - a single number - the intended size of each batches

#What do need to implement
# 1) Shuffle data
# - Gotta shuffle X and y the same way not to break the correspondence between X_i and y_i
# 3) Split data into minibatches of batch_size
# - If data size is not a multiple of batch_size, make one last batch smaller.
# 4) return a list (or an iterator) of pairs
# - ( ,   y   )
def iterate_minibatches(X, y, batchsize):
    assert len(X) == len(y)
    indices = np.arange(len(X))
    np.random.shuffle(indices)
    for start_idx in range(0, len(X) - batchsize + 1, batchsize):
        excerpt = indices[start_idx:start_idx + batchsize]
        yield X[excerpt], y[excerpt]
    #<return an iterable of (X_batch, y_batch)  batches of images and answers for them>
````

# Training loop

````python
import time

num_epochs = 100 #amount of passes through the data
            
batch_size = 50 #number of samples processed at each function call

for epoch in range(num_epochs):
    # In each epoch, we do a full pass over the training data:
    train_err = 0
    train_acc = 0
    train_batches = 0
    start_time = time.time()
    for batch in iterate_minibatches(X_train, y_train, batch_size):
        inputs, targets = batch
        train_err_batch, train_acc_batch = train_fun(inputs, targets)
        train_err += train_err_batch
        train_acc += train_acc_batch
        train_batches += 1

    # And a full pass over the validation data:
    val_acc = 0
    val_batches = 0
    for batch in iterate_minibatches(X_val, y_val, batch_size):
        inputs, targets = batch
        val_acc += accuracy_fun(inputs, targets)
        val_batches += 1

    # Then we print the results for this epoch:
    print("Epoch {} of {} took {:.3f}s".format(epoch + 1, num_epochs, time.time() - start_time))
    print("  training loss (in-iteration):\t\t{:.6f}".format(train_err / train_batches))
    print("  train accuracy:\t\t{:.2f} %".format(train_acc / train_batches * 100))
    print("  validation accuracy:\t\t{:.2f} %".format(val_acc / val_batches * 100))
````

````python
test_acc = 0
test_batches = 0
for batch in iterate_minibatches(X_test, y_test, 500):
    inputs, targets = batch
    test_acc += accuracy_fun(inputs, targets)
    test_batches += 1
print("Final results:")
print("  test accuracy:\t\t{:.2f} %".format(test_acc / test_batches * 100))

if test_acc / test_batches * 100 > 95:
    print "Double-check, than consider applying for NIPS'17. SRSly."
elif test_acc / test_batches * 100 > 90:
    print "U'r freakin' amazin'!"
elif test_acc / test_batches * 100 > 80:
    print "Achievement unlocked: 110lvl Warlock!"
elif test_acc / test_batches * 100 > 70:
    print "Achievement unlocked: 80lvl Warlock!"
elif test_acc / test_batches * 100 > 50:
    print "Achievement unlocked: 60lvl Warlock!"
else:
    print "We need more magic!"
````

# First step

````python
import numpy as np
from cifar import load_cifar10
X_train,y_train,X_val,y_val,X_test,y_test = load_cifar10("./new_cifar/")

class_names = np.array(['airplane','automobile ','bird ','cat ','deer ','dog ','frog ','horse ','ship ','truck'])

print X_train.shape,y_train.shape
````

````python
import lasagne
import theano
import theano.tensor as T

input_X = T.tensor4("X")

#input dimention (None means "Arbitrary")
input_shape = [None,3,32,32]

target_y = T.vector("target Y integer",dtype='int32')
````

Defining network architecture

````python
#Input layer (auxilary)
input_layer = lasagne.layers.InputLayer(shape=input_shape, input_var=input_X,
                                        name='input')

#3x3 convolution with 10 filters and ReLU activation
conv = lasagne.layers.Conv2DLayer(input_layer, num_filters=10, filter_size=(3,3), 
                                  nonlinearity = lasagne.nonlinearities.rectify,
                                  name='conv')

#3x3 pooling
pool = lasagne.layers.Pool2DLayer(conv, pool_size=(3,3), mode='max', name='pool')

#Dense layer with 100-neurons and ReLU activation
dense = lasagne.layers.DenseLayer(pool, num_units=100,
                                  nonlinearity=lasagne.nonlinearities.rectify,
                                  name='dense')

#10% dropout
drop = lasagne.layers.DropoutLayer(dense, p=0.1, name='drop')

#Output dense layer
dense_output = lasagne.layers.DenseLayer(drop, num_units = 10,
                                         nonlinearity = lasagne.nonlinearities.softmax,
                                         name='output')
````

````python
#network prediction (theano-transformation)
y_predicted = lasagne.layers.get_output(dense_output)
````

````python
#all network weights (shared variables)
all_weights = lasagne.layers.get_all_params(dense_output,trainable=True)
print all_weights
````

````python
#Mean categorical crossentropy as a loss function - similar to logistic loss but for multiclass targets
loss = lasagne.objectives.categorical_crossentropy(y_predicted,target_y).mean()

#prediction accuracy (WITH dropout)
accuracy = lasagne.objectives.categorical_accuracy(y_predicted,target_y).mean()

#This function computes gradient AND composes weight updates just like you did earlier
updates_sgd = lasagne.updates.adam(loss, all_weights,learning_rate=0.01)
````

````python
#function that computes loss and updates weights
train_fun = theano.function([input_X,target_y],[loss,accuracy], updates=updates_sgd)
````

````python
#deterministic prediciton (without dropout)
y_predicted_det = lasagne.layers.get_output(dense_output, deterministic=True)

#prediction accuracy (without dropout)
accuracy_det = lasagne.objectives.categorical_accuracy(y_predicted_det, target_y).mean()

#function that just computes accuracy without dropout/noize -- for evaluation purposes
accuracy_fun = theano.function([input_X,target_y], accuracy_det)
````

Training loop

````python
import time

num_epochs = 10 #amount of passes through the data
            
batch_size = 50 #number of samples processed at each function call

for epoch in range(num_epochs):
    # In each epoch, we do a full pass over the training data:
    train_err = 0
    train_acc = 0
    train_batches = 0
    start_time = time.time()
    for batch in iterate_minibatches(X_train, y_train, batch_size):
        inputs, targets = batch
        train_err_batch, train_acc_batch = train_fun(inputs, targets)
        train_err += train_err_batch
        train_acc += train_acc_batch
        train_batches += 1

    # And a full pass over the validation data:
    val_acc = 0
    val_batches = 0
    for batch in iterate_minibatches(X_val, y_val, batch_size):
        inputs, targets = batch
        val_acc += accuracy_fun(inputs, targets)
        val_batches += 1

    # Then we print the results for this epoch:
    print("Epoch {} of {} took {:.3f}s".format(epoch + 1, num_epochs, time.time() - start_time))
    print("  training loss (in-iteration):\t\t{:.6f}".format(train_err / train_batches))
    print("  train accuracy:\t\t{:.2f} %".format(train_acc / train_batches * 100))
    print("  validation accuracy:\t\t{:.2f} %".format(val_acc / val_batches * 100))
````

````python
test_acc = 0
test_batches = 0
for batch in iterate_minibatches(X_test, y_test, 500):
    inputs, targets = batch
    test_acc += accuracy_fun(inputs, targets)
    test_batches += 1
print("Final results:")
print("  test accuracy:\t\t{:.2f} %".format(test_acc / test_batches * 100))

if test_acc / test_batches * 100 > 95:
    print "Double-check, than consider applying for NIPS'17. SRSly."
elif test_acc / test_batches * 100 > 90:
    print "U'r freakin' amazin'!"
elif test_acc / test_batches * 100 > 80:
    print "Achievement unlocked: 110lvl Warlock!"
elif test_acc / test_batches * 100 > 70:
    print "Achievement unlocked: 80lvl Warlock!"
elif test_acc / test_batches * 100 > 50:
    print "Achievement unlocked: 60lvl Warlock!"
else:
    print "We need more magic!"
````

## Tips on what can be done:


 * __Network size__
   * MOAR neurons, 
   * MOAR layers, ([lasagne docs](http://lasagne.readthedocs.org))

   * Nonlinearities in the hidden layers
     * tanh, relu, leaky relu, etc
   * Larger networks may take more epochs to train, so don't discard your net just because it could didn't beat the baseline in 5 epochs.

   * Ph'nglui mglw'nafh Cthulhu R'lyeh wgah'nagl fhtagn!

 * __Convolution layers__
   * they __are a must__ unless you have any super-ideas
   * `network = lasagne.layers.Conv2DLayer(prev_layer,`
    `                        num_filters = n_neurons,`
    `                        filter_size = (filter width, filter height),`
    `                        nonlinearity = some_nonlinearity)`
   * Warning! Training convolutional networks can take long without GPU. That's okay.
     * If you are CPU-only, we still recomment to try a simple convolutional architecture
     * a perfect option is if you can set it up to run at nighttime and check it up at the morning.
     * Make reasonable layer size estimates. A 128-neuron first convolution is likely an overkill.
     * __To reduce computation__ time by a factor in exchange for some accuracy drop, try using __stride__ parameter. A stride=2 convolution should take roughly 1/4 of the default (stride=1) one.
 
   * Plenty other layers and architectures
     * http://lasagne.readthedocs.org/en/latest/modules/layers.html
     * batch normalization, pooling, etc

 * __Early Stopping__
   * Training for 100 epochs regardless of anything is probably a bad idea.
   * Some networks converge over 5 epochs, others - over 500.
   * Way to go: stop when validation score is 10 iterations past maximum
     

 * __Faster optimization__ - 
   * rmsprop, nesterov_momentum, adam, adagrad and so on.
     * Converge faster and sometimes reach better optima
     * It might make sense to tweak learning rate/momentum, other learning parameters, batch size and number of epochs
   * __BatchNormalization__ (lasagne.layers.batch_norm) FTW!


 * __Regularize__ to prevent overfitting
   * Add some L2 weight norm to the loss function, theano will do the rest
     * Can be done manually or via - http://lasagne.readthedocs.org/en/latest/modules/regularization.html
   * Dropout - to prevent overfitting
     * `lasagne.layers.DropoutLayer(prev_layer, p=probability_to_zero_out)`   
     * Don't overdo it. Check if it actually makes your network better
   
   
 * __Data augmemntation__ - getting 5x as large dataset for free is a great deal
   * Zoom-in+slice = move
   * Rotate+zoom(to remove black stripes)
   * any other perturbations
   * Add Noize (easiest: GaussianNoizeLayer)
   * Simple way to do that (if you have PIL/Image): 
     * ```from scipy.misc import imrotate,imresize```
     * and a few slicing
   * Stay realistic. There's usually no point in flipping dogs upside down as that is not the way you usually see them.

````python
import numpy as np
from cifar import load_cifar10
X_train,y_train,X_val,y_val,X_test,y_test = load_cifar10("./new_cifar/")

class_names = np.array(['airplane','automobile ','bird ','cat ','deer ','dog ','frog ','horse ','ship ','truck'])

print X_train.shape,y_train.shape
````

````python
import lasagne
import theano
import theano.tensor as T

input_X = T.tensor4("X")

#input dimention (None means "Arbitrary")
input_shape = [None,3,32,32]

target_y = T.vector("target Y integer",dtype='int32')
````

Defining network architecture

````python
#Input layer (auxilary)
input_layer = lasagne.layers.InputLayer(shape=input_shape, input_var=input_X,
                                        name='input')

#3x3 convolution with 16 filters and ReLU activation, uniform init
conv1 = lasagne.layers.Conv2DLayer(input_layer, num_filters=16, filter_size=(3,3), 
                                  nonlinearity = lasagne.nonlinearities.rectify,
                                  name='conv1')

#batch_norm (with default params)
batch1 = lasagne.layers.batch_norm(conv1, name='batch1')

#3x3 convolution with 16 filters and ReLU activation, uniform init
conv2 = lasagne.layers.Conv2DLayer(batch1, num_filters=16, filter_size=(3,3), 
                                  nonlinearity = lasagne.nonlinearities.rectify,
                                  name='conv2')

#3x3 pooling
pool1 = lasagne.layers.Pool2DLayer(conv2, pool_size=(3,3), mode='max', name='pool1')

#3x3 convolution with 32 filters and ReLU activation, uniform init
conv3 = lasagne.layers.Conv2DLayer(pool1, num_filters=32, filter_size=(3,3), 
                                  nonlinearity = lasagne.nonlinearities.rectify,
                                  name='conv3')

#batch_norm (with default params)
batch2 = lasagne.layers.batch_norm(conv3, name='batch2')

#3x3 convolution with 32 filters and ReLU activation, uniform init
conv4 = lasagne.layers.Conv2DLayer(batch2, num_filters=32, filter_size=(3,3), 
                                  nonlinearity = lasagne.nonlinearities.rectify,
                                  name='conv4')

#3x3 pooling
pool2 = lasagne.layers.Pool2DLayer(conv4, pool_size=(3,3), mode='max', name='pool2')

#Dense layer with 128-neurons and ReLU activation
dense1 = lasagne.layers.DenseLayer(pool2, num_units=128,
                                  nonlinearity=lasagne.nonlinearities.rectify,
                                  name='dense1')

#10% dropout
drop1 = lasagne.layers.DropoutLayer(dense1, p=0.1, name='drop1')

#Dense layer with 128-neurons and ReLU activation
dense2 = lasagne.layers.DenseLayer(drop1, num_units=128,
                                  nonlinearity=lasagne.nonlinearities.rectify,
                                  name='dense2')

#10% dropout
drop2 = lasagne.layers.DropoutLayer(dense2, p=0.1, name='drop2')

#Output dense layer
dense_output = lasagne.layers.DenseLayer(drop2, num_units = 10,
                                         nonlinearity = lasagne.nonlinearities.softmax,
                                         name='output')
````

````python
# Network predictions (theano-transformation)
y_predicted = lasagne.layers.get_output(dense_output)
````

````python
#All weights (shared-varaibles)
# "trainable" flag means not to return auxilary params like batch mean (for batch normalization)
all_weights = lasagne.layers.get_all_params(dense_output,trainable=True)
print all_weights
````

````python
#Mean categorical crossentropy as a loss function - similar to logistic loss but for multiclass targets
loss = lasagne.objectives.categorical_crossentropy(y_predicted,target_y).mean()

#prediction accuracy (WITH dropout)
accuracy = lasagne.objectives.categorical_accuracy(y_predicted,target_y).mean()

#This function computes gradient AND composes weight updates just like you did earlier
updates_sgd = lasagne.updates.adam(loss, all_weights,learning_rate=0.01)
````

````python
#A function that accepts X and y, returns loss functions and performs weight updates
train_fun = theano.function([input_X,target_y], [loss,accuracy], updates=updates_sgd)
````

````python
#deterministic prediciton (without dropout)
y_predicted_det = lasagne.layers.get_output(dense_output)

#prediction accuracy (without dropout)
accuracy_det = lasagne.objectives.categorical_accuracy(y_predicted_det,target_y).mean()

#function that just computes accuracy without dropout/noize -- for evaluation purposes
accuracy_fun = theano.function([input_X,target_y],accuracy_det)
````

In the first and second steps I tried to use a mini-convolutional network. The only difference between these two networks was about using batch_norm between convolution and pooling. To just have a look on these two networks I trained them for only 10 iterations, but the second one (with batch_norm) had already seemed significantly better.

Next, in the first attempt I wanted to improve this architecture by adding another *conv-batch-pool* chain and *dense-drop* chain and also increasing the number of neurons. I have included the early stopping check in the training loop, so now the training procedure have stopped if the validation score is less than maximum for 10 iterations. For training I used the same Adam algo with learning rate 0.01. Eventually, this network "converged" over 37 epochs and it gave 62% accuracy on test.

Further, in the second attempt I tried to modify a little the previous network. I have just put two additioanal *conv* both before two *pool*. I did not change anything else, but it turned out that it gave only 51% accuracy on test data and took more than 7 hours for "convergence" on my machine. So another improvments would take even more time.

To sum up, the final best architecture in my case is *input-conv16(3x3)-batch-pool(3x3)-conv32(3x3)-batch-pool(3x3)-dense128-drop10%-dense128-drop10%-output*. It gave 65% max accuracy on training, 62% max accuracy on validation and 62% accuracy on test. I think that additional *conv* before *pool* made results worse because it reduced the images too much. Probably, it would be better to make 2x2 pool instead 3x3 in this case to compensate the images' sizes.
